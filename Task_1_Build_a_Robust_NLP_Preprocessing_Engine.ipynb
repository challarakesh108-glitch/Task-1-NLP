#Task 1: Conceptual Understanding

1.“Love” and “love” are different because of uppercase vs lowercase.Their meaning is same, but programmatically they are processed as two separate words unless text is converted to lowercase.

2.Text becomes long and noisy,Model processing time increases,Important words get less weight,Accuracy of classification or clustering may reduce

3.Sentiment Analysis:Removing “not” changes the meaning:“I am not happy” → “I am happy”

Legal / Medical DocumentsWords like “is”, “not”, “no”, “under” matter:“Patient is not responding.”“He is not guilty.”


4.stemming : Rough cutting of words,Not grammatically correct,Fast, rule-based,Example: “studies → studi”  
 Lemmatization : Uses dictionary meaning,Grammatically valid output,Slow, meaning-based,example:“studies → study”


```python
#Task 2:Build Advanced Preprocessing Function
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return []

    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters/emojis
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short words except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return tokens
```

```python
#Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned_sentence)
```

```python
#Task 4: Token Analytics
def analyze_tokens(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    total, unique, avg = analyze_tokens(tokens)

    print(f"\nSentence: {sentence}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Avg Token Length: {avg}")
```

```python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

counter = Counter(all_tokens)

print("Top 10 frequent words:", counter.most_common(10))

print("Least frequent words:", counter.most_common()[-5:])
```

```python
#Task 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text) # preprocess_text returns only tokens
        cleaned = " ".join(tokens) # Create the cleaned string from the tokens
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)
output
```

```python
#Task 7: Error Handling
edge_cases = [
    "",
    "😂😂😂😂",
    "123456789",
    "!!!!!!!"
]

for case in edge_cases:
    print("\nInput:", case)
    print("Output:", preprocess_text(case))
```
